```python
def showtimeserie(batch, filename, activity, subject):
    #Data broad graphic
    batch_wide = batch
    
    #range
    batch_range = batch_wide.shape[0]
    
    x_values = list(range(0,batch_range))
    

    y_value1 = batch_wide['magnitude_chest']
    y_value2 = batch_wide['magnitude_forearm']
    y_value3 = batch_wide['magnitude_head']
    y_value4 = batch_wide['magnitude_shin']
    y_value5 = batch_wide['magnitude_thigh']
    y_value6 = batch_wide['magnitude_upperarm']
    y_value7 = batch_wide['magnitude_waist']
    
    # multiple line plots
    plt.plot( x_values, y_value1, marker='', markerfacecolor='blue', color='skyblue', linewidth=2, label="mag_chest")
    plt.plot( x_values, y_value2, marker='', color='red', linewidth=2, label="mag_forearm")
    plt.plot( x_values, y_value3, marker='', color='olive', linewidth=2, label="mag_head")
    plt.plot( x_values, y_value4, marker='', color='blue', linewidth=2, label="mag_shin")
    plt.plot( x_values, y_value5, marker='', color='green', linewidth=2, label="mag_thigh")
    plt.plot( x_values, y_value6, marker='', color='black', linewidth=2, label="mag_upperarm")
    plt.plot( x_values, y_value7, marker='', color='purple', linewidth=2, label="mag_waist")

    # show legend
    plt.legend()

    #other
    plt.rcParams["figure.figsize"] = [20, 15]
    plt.title(str(activity) + '- Subject ' + str(subject) + ' - Batch['+ str(batch_range) + '] - magnitude"s accelerometer data')

    #save graph
    plt.savefig(filename)

    # show graph
    plt.show()
```

```python
#climbingdown
batch = df_try.loc[(df_try['Activity'] == 'climbingdown') & (df_try['Subject ID'] == 10)]
showtimeserie(batch, 'timeserie_climbingdown_10.png', 'climbingdown', 10)
```

```python
#climbingup
batch2 = df_try.loc[(df_try['Activity'] == 'climbingup') & (df_try['Subject ID'] == 10)]
showtimeserie(batch2, 'timeserie_climbingup_10.png', 'climbingup', 10)

#running
batch3 = df_try.loc[(df_try['Activity'] == 'running') & (df_try['Subject ID'] == 10)]
showtimeserie(batch3, 'timeserie_running_10.png', 'running', 10)

#jumping
batch4 = df_try.loc[(df_try['Activity'] == 'jumping') & (df_try['Subject ID'] == 10)]
showtimeserie(batch4, 'timeserie_jumping_10.png', 'jumping', 10)

#walking
batch5 = df_try.loc[(df_try['Activity'] == 'walking') & (df_try['Subject ID'] == 10)]
showtimeserie(batch5, 'timeserie_walking_10.png', 'walking', 10)

#lying
batch6 = df_try.loc[(df_try['Activity'] == 'lying') & (df_try['Subject ID'] == 10)]
showtimeserie(batch6, 'timeserie_lying_10.png', 'lying', 10)

#standing
batch7 = df_try.loc[(df_try['Activity'] == 'standing') & (df_try['Subject ID'] == 10)]
showtimeserie(batch7, 'timeserie_standing_10.png', 'standing', 10)

#sitting
batch8 = df_try.loc[(df_try['Activity'] == 'sitting') & (df_try['Subject ID'] == 10)]
showtimeserie(batch8, 'timeserie_sitting_10.png', 'sitting', 10)
```
