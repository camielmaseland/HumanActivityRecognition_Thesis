```python
#create index to merge on
df_acc = df_acc.reset_index()
df_gyr = df_gyr.reset_index()
df_mag = df_mag.reset_index()
```

```python
# Add suffixes to dataframes
df_acc = df_acc.add_suffix('_acc')
df_gyr = df_gyr.add_suffix('_gyr')
df_mag = df_mag.add_suffix('_mag')
```

### Merge acc, gyr and mag csv files

```python
final_set = df_acc
final_set.shape
```

```python
final_set = final_set.merge(df_gyr, how='inner', 
                            left_on=["id_acc", "Subject ID_acc", "Activity_acc"], 
                            right_on=["id_gyr", "Subject ID_gyr", "Activity_gyr"])
print('Merge Acc + Gyr: ' + str(final_set.shape))
final_set
```

```python
final_set = final_set.merge(df_mag, how='inner', 
                            left_on=["id_acc", "Subject ID_acc", "Activity_acc"], 
                            right_on=["id_mag", "Subject ID_mag", "Activity_mag"])
print('Merge Acc + Gyr + Mag: ' + str(final_set.shape))
final_set
```

### Reorder columns
Identifying columns to the back

```python
#first drop duplicate columns
final_set = final_set.drop(columns=['id_gyr', 'Subject ID_gyr', 'Activity_gyr', 'id_mag', 'Subject ID_mag', 'Activity_mag',
                                   'index_gyr', 'index_mag'])

#rename identifying columns back to normal
final_set.rename(columns={'id_acc': 'id', 'Subject ID_acc': 'Subject ID', 'Activity_acc': 'Activity'}, inplace=True)
```

```python
#reorder such that labels are at the back
final_set = final_set[[c for c in final_set if c not in ['Subject ID', 'Activity']] 
                      + ['Subject ID', 'Activity']]
```

### Label encoder - Activity

- 0 = climbingdown
- 1 = climbingup
- 2 = jumping
- 3 = lying
- 4 = running
- 5 = sitting
- 6 = standing
- 7 = walking

```python
from sklearn import preprocessing

#imablance before encode
bef_enc = Counter(final_set['Activity'])
        
#Encode Activity
le = preprocessing.LabelEncoder()
final_set['Activity'] = le.fit_transform(final_set.Activity.values)
```

```python
#check imbalance
aft_enc = Counter(final_set['Activity'])

print(bef_enc)
print(aft_enc)
```
